```python
df_2 = df_1.rename(columns={'Question 13: 13. Which option best describes your feelings towards vaccines?': 'Sentiment'})
```

```python
df_2
```

```python
df_2.loc[df_2['Sentiment']=='Positive', 'sentiment_encoded'] = 2
df_2.loc[df_2['Sentiment']=='Negative', 'sentiment_encoded'] = 1
df_2.loc[df_2['Sentiment']=='Neutral', 'sentiment_encoded'] = 0
df_2.loc[df_2['Sentiment']=='Very positive', 'sentiment_encoded'] = 3

```

```python
df_2
```

```python
df_2['Sentiment'].value_counts()
```

```python
sentiment_overall_values = [22, 14, 9, 2]
sentiment_overall_category = ['Very positive', 'Positive', 'Neutral', 'Negative']
```

```python
import plotly.express as px
```

```python
fig = px.bar(x = sentiment_overall_category, y = sentiment_overall_values, text_auto = True,
            color = sentiment_overall_category,
             color_discrete_map = {'Neutral': 'blue', 
                                   'Positive': 'green', 
                                   'Negative': 'red',
                                  'Very positive': 'darkgreen'})
fig.update_traces(marker_line_color = 'black',
                  marker_line_width = 2, opacity = 1)
fig.update_layout(
    title=dict(text="Vaccine Survey Sentiment", font=dict(size=30))
)
```
