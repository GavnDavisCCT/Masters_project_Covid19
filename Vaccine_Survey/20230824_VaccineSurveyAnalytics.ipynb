```python
sentiment_overall_values = [22, 14, 9, 2]
sentiment_overall_category = ['Very positive', 'Positive', 'Neutral', 'Negative']
```

```python
import plotly.express as px
```

```python
fig = px.bar(x = sentiment_overall_category, y = sentiment_overall_values, text_auto = True,
            color = sentiment_overall_category,
             color_discrete_map = {'Neutral': 'blue', 
                                   'Positive': 'green', 
                                   'Negative': 'red',
                                  'Very positive': 'darkgreen'})
fig.update_traces(marker_line_color = 'black',
                  marker_line_width = 2, opacity = 1)
fig.update_layout(
    title=dict(text="Vaccine Survey Sentiment", font=dict(size=30))
)
```

```python
x = ['18-24', '25-34', '35-50', '51-65']
y1 = np.array([0, 0, 1, 0])
y2 = np.array([12, 9, 5, 1])
y3 = np.array([8, 3, 2, 1])
y4 = np.array([2, 2, 1, 0])

plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.bar(x, y3, bottom=y1+y2, color='y')
plt.bar(x, y4, bottom=y1+y2+y3, color='g')
plt.xlabel("Age Bracket")
plt.ylabel("Count")
plt.legend(["Negative", "Very positive", "positve", "neutral"])
plt.title("Scores by Teams in 4 Rounds")
plt.show()
```
